```python
import panel as pn
import plotly.express as px
import pandas as pd

pn.extension('plotly')

df = pd.read_excel("24A_jelentkezok_kepzesenkent.xlsx")
fig = px.sunburst(df, path=['fenntarto', 'intezmeny', 'kar'], values='elsohelyes_jelentkezo')


# Create a Panel object to hold the Plotly figure
plot_pane = pn.pane.Plotly(fig, sizing_mode='stretch_both')

# Create a responsive layout with a title and the Plotly figure
dashboard = pn.Column(
    "# Panel Dashboard Example",
    plot_pane,
    sizing_mode='stretch_both'
)

# Save the layout to an HTML file
dashboard.save('index.html', embed=True)
```

```python
df = pd.read_excel("24A_jelentkezok_kepzesenkent.xlsx")

#drop cases where no-one applied
df = df[df['elsohelyes_jelentkezo'] > 0]  

custom_color_scale = [
    [0, 'red'],
    [0.5, 'yellow'],
    [1, 'green']
]

fig = px.sunburst(df, 
    # path=['fenntarto', 'intezmeny', 'kar', 'kepzesi_terulet', 'mkpz_nev'], 
    path=['fenntarto', 'intezmeny', 'kar', 'kepzesi_terulet'], 
    values='elsohelyes_jelentkezo',
    color='elsohelyes_jelentkezo',

    color_continuous_scale=custom_color_scale,
    )
fig.show()

#If a color argument is passed, the color of a node is computed as the average of the color values of its children, weighted by their values.
```
